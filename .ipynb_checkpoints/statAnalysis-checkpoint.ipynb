```python
df = pd.read_csv('../train.csv')
df_test = pd.read_csv('../test.csv')

# print(df.columns.values) uncomment to see the 
```

```python
df = df[['id','target','comment_text']]
```

```python
# sequence length
# toxicity
# out_of_words
```

```python
vocab_file = '../uncased_L-12_H-768_A-12/vocab.txt'
      
tokenizer = tokenization.FullTokenizer(vocab_file=vocab_file, do_lower_case=False)
df['tokenized'] = df["comment_text"].apply(lambda x: tokenizer.tokenize(x))
```

```python
tknzr = TweetTokenizer()
df['twitter_tokenized'] = df["comment_text"].apply(lambda x: tknzr.tokenize(x))
```

```python
df.to_csv("tokenizen_train")
```

```python
df['toxic_class'] = df["target"].apply(lambda x: 1 if x > 0 else 0)
df['raw_length'] = df["comment_text"].apply(lambda x: len(x))
df['token_length'] = df["tokenized"].apply(lambda x: len(x))
df['tweet_token_length'] = df["twitter_tokenized"].apply(lambda x: len(x))
df['length_diff'] = df['token_length'].sub(df['tweet_token_length'], axis = 0) 
df['unique_tokens'] = df["tokenized"].apply(lambda x: len(set(x)))
```

```python
import seaborn as sns
x = pd.Series(df["token_length"])
ax = sns.distplot(x, kde=False)
```
